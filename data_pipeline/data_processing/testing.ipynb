```python
import matplotlib.pyplot as plt
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from numpy import ndarray, dtype, float64

output_notebook()
```

```python

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors: float | ndarray[tuple[int, ...], dtype[float64]] = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python

N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = np.array([(r, g, 150) for r, g in zip(50 + 2 * x, 30 + 2 * y)], dtype="uint8")

TOOLS = "hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,examine,help"

p = figure(tools=TOOLS)

p.circle(x, y, radius=radii,
         fill_color=colors, fill_alpha=0.6,
         line_color=None)

show(p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Define a function to plot
def plot_function(amplitude, frequency, phase):
    x = np.linspace(0, 10, 1000)
    y = amplitude * np.sin(frequency * x + phase)

    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=f'Amplitude = {amplitude}, Frequency = {frequency}, Phase = {phase}')
    plt.title("Interactive Sine Wave")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.ylim(-10, 10)
    plt.legend()
    plt.grid()
    plt.show()

# Add sliders for amplitude, frequency, and phase
interact(
    plot_function,
    amplitude=FloatSlider(min=1, max=10, step=0.5, value=5, description='Amplitude:'),
    frequency=FloatSlider(min=1, max=5, step=0.1, value=2, description='Frequency:'),
    phase=FloatSlider(min=0, max=2*np.pi, step=0.1, value=0, description='Phase:')
)
```
